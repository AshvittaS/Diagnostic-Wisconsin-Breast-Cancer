```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
cols = [
    "ID", "Diagnosis",
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean",
    "compactness_mean", "concavity_mean", "concave_points_mean", "symmetry_mean",
    "fractal_dimension_mean",
    "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se",
    "compactness_se", "concavity_se", "concave_points_se", "symmetry_se",
    "fractal_dimension_se",
    "radius_worst", "texture_worst", "perimeter_worst", "area_worst", "smoothness_worst",
    "compactness_worst", "concavity_worst", "concave_points_worst", "symmetry_worst",
    "fractal_dimension_worst"
]
df = pd.read_csv(url, header=None, names=cols)
print(df.head())
```

```python
#Based on correlation matrix we can drop some features
corr_matrix = df.corr().abs()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)
drop_Col = [x for x in tri_df.columns if any(tri_df[x]>0.90)]
df = df.drop(drop_Col, axis=1)
print(df.shape[1])
```

```python
Y = df["Diagnosis"]
X = df.drop('Diagnosis', axis=1)
```

```python
X.head()
```

```python
skewness = X.skew()
skewed_cols = skewness[skewness > 0.1].index.tolist()
for col in skewed_cols:
    if (X[col] <= 0).any():
        X[col] = X[col] + abs(X[col].min()) + 1
    X[col] = np.log(X[col])
```

```python
len(X.columns)
```

```python
len(skewed_cols)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled.shape
```

```python
import pickle
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

```python
rows = df.drop('Diagnosis', axis=1).columns
rows.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_scaled)
plt.plot(range(1, X_scaled.shape[1] + 1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs Number of Components')
plt.grid(True)
plt.show()
```

```python
n_components = 6
pca = PCA(n_components=n_components)
components=pca.fit_transform(X_scaled)
```

```python
with open("pca.pkl", "wb") as f:
    pickle.dump(pca, f)
```

```python
columns = []
for i in range(n_components):
  columns.append(f'PC_{i+1}')
```

```python
columns
```

```python
feature_importance=pd.DataFrame(pca.components_.T,index=rows,columns=columns)
feature_importance
```

```python
total_var=pca.explained_variance_ratio_.sum()*100
total_var
```

```python
variance_ratios = pd.DataFrame(pca.explained_variance_ratio_.reshape(1,n_components), columns=columns, index=['Explained Variance Ratio'])
variance_ratios
```

# Visualisation

```python
plt.figure(figsize=(15, 14))
sns.heatmap(feature_importance,annot=True)
plt.title('Principal Components correlation with the features')
plt.xlabel('Principal Components')
plt.ylabel('Features')
plt.show()
```

```python
import plotly.express as px
x_labels = {
   # str(i): f"PC{i+1}"
   # for i in range(n_components)
    str(i): f"PC {i+1} ({var:.1f}%)"for i, var in enumerate(pca.explained_variance_ratio_ * 100)}

y_labels = {str(i): f"PC{i+1}"for i in range(n_components)}
fig = px.scatter_matrix(
    components,
    color=df['Diagnosis'],
    dimensions=range(n_components),
    labels=y_labels,
    title=f'Total Explained Variance: {total_var:.2f}%',
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(components,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
with open("logreg.pkl", "wb") as f:
    pickle.dump(lr, f)
```

```python
y_pred_test=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,RocCurveDisplay
print(
    f"Classification report for classifier {lr}:\n"
    f"{classification_report(y_test, y_pred_test)}\n"
)
```

```python
co=confusion_matrix(y_test,y_pred_test)
```

```python
sns.heatmap(co,annot=True)
```

```python
print(accuracy_score(y_pred_test,y_test))
```
